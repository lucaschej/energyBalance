$Q_s=-Q_f$

$k_s \nabla T_s=-k_f \nabla T_f$

$k_s\frac{(Tw-Ts)}{y_s}=-k_f\frac{(Tw-Tf)}{y_f}$

$y_f = \frac{k_f}{k_s}\frac{(T_f-T_w)}{(T_w-T_s)}y_s$

$y_f = \frac{0.6 W/mK}{0.2 W/mK}\frac{(T_f-T_w)}{(T_w-T_s)}1$

$y_f = 3 \frac{(T_f-T_w)}{(T_w-T_s)}$

$b=\frac{(T_f-T_w)}{(T_w-T_s)}$

$y_f = 3 b$   con   $0<b<1$

A mayor diferencia entre $(T_w-T_s)$, menor será $b$, y menor tendrá que ser $y_f$ para poder calcular bien la transferencia de calor.

```python
import numpy as np
import matplotlib.pyplot as plt

def y_fluid(b):
    return 3 * b

# Generamos muchos puntos entre 0 y 1
num_points = 1000
b_values = np.linspace(0, 1, num_points)

# Calculamos los valores de y_fluid para cada valor de b
y_values = y_fluid(b_values)

# Graficamos la función
plt.plot(b_values, y_values)
plt.axhline(y=1, color='r', linestyle='--', linewidth=1)  # Línea horizontal en y=1
plt.xlabel('b = Tf-Tw / Tw - Ts')
plt.ylabel('y_fluid')
plt.title('Gráfico de y_fluid = 3*b')
plt.gca().invert_xaxis()  # Invertimos el eje x para que vaya de 1 a 0
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos proporcionados
time = np.array([0, 1, 2, 3, 4])
power = np.array([0, -0.003516782, -0.005537041, -0.006899228, -0.00785769])

# Función exponencial decreciente para el ajuste
def exponential_function(t, a, b, c):
    return a * np.exp(-b * t) + c

# Ajuste de la curva a los datos
params, _ = curve_fit(exponential_function, time, power)

# Generar valores de tiempo para predecir hasta 60 segundos
time_predict = np.linspace(0, 60, 100)

# Predecir los valores de potencia usando la función ajustada
power_predict = exponential_function(time_predict, *params)

# Graficar los datos originales y la función ajustada
plt.plot(time, power, marker='o', label='Datos originales')
plt.plot(time_predict, power_predict, label='Ajuste exponencial')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Potencia')
plt.legend()
plt.grid(True)
plt.title('Ajuste exponencial de la curva')
plt.show()
```
